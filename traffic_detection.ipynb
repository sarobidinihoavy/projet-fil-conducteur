First, let's perform the validation step to gather all the evaluation metrics and ensure the latest results are available.

```
from ultralytics import YOLO

# Load the best trained model
model = YOLO('/content/runs/detect/train/weights/best.pt')

# Validate the model on the test dataset to get evaluation metrics
# The default data='data.yaml' will be used if not specified, which was used during training.
# If you wish to use a different dataset for validation, you can specify it here.
metrics = model.val(plots=True) # plots=True ensures plots are generated/regenerated
```

Now, let's display the key evaluation metrics from the validation run.

```
print("--- Model Evaluation Metrics ---")
print(f"Map (Mean Average Precision) at IoU 0.5-0.95: {metrics.box.map:.3f}")
print(f"Map (Mean Average Precision) at IoU 0.50: {metrics.box.map50:.3f}")
print(f"Map (Mean Average Precision) at IoU 0.75: {metrics.box.map75:.3f}")

# You can also access metrics per class
class_names = metrics.names
class_maps = metrics.box.maps # mAP for each class

print("\n--- mAP per Class ---")
for i, name in enumerate(class_names):
    print(f"  {name}: {class_maps[i]:.3f}")

# You can also get precision and recall values if needed, though they are usually part of curves.
# For example, metrics.box.mean_precision and metrics.box.mean_recall could be accessed
# if the `metrics` object provided direct attributes for them, otherwise they are derived from PR curves.
```

Ultralytics automatically generates several plots that visualize the model's performance. These are typically saved in the `runs/detect/val/` directory (or `runs/detect/train/` if `val()` is run immediately after `train()` without clearing the directory). Let's list the generated plots and display a few important ones.

```
import os
from IPython.display import Image, display

# The validation results are saved in a directory like runs/detect/val, potentially with a run number
# Let's find the latest validation run directory
val_runs_dir = '/content/runs/detect/'
if os.path.exists(val_runs_dir):
    all_runs = sorted([d for d in os.listdir(val_runs_dir) if os.path.isdir(os.path.join(val_runs_dir, d))], reverse=True)
    latest_val_dir = None
    for run_dir in all_runs:
        # Check for directories that start with 'val' (from val mode) or 'train' (from train mode validation)
        if run_dir.startswith('val') or run_dir.startswith('train'):
            latest_val_dir = os.path.join(val_runs_dir, run_dir)
            break

    if latest_val_dir and os.path.exists(latest_val_dir):
        print(f"Plots found in: {latest_val_dir}")
        plot_files = [f for f in os.listdir(latest_val_dir) if f.endswith('.png') or f.endswith('.jpg')]
        print("\n--- Available Plots ---")
        for f in plot_files:
            print(f"- {f}")

        # Display some important plots
        print("\n--- Displaying Key Plots ---")
        key_plots = ['confusion_matrix.png', 'BoxF1_curve.png', 'BoxP_curve.png', 'BoxR_curve.png'] # Corrected filenames

        for plot_name in key_plots:
            plot_path = os.path.join(latest_val_dir, plot_name)
            if os.path.exists(plot_path):
                print(f"\nDisplaying {plot_name}:")
                display(Image(filename=plot_path, width=800))
            else:
                print(f"Plot '{plot_name}' not found at {plot_path}")
    else:
        print("No validation run directory found to display plots.")
else:
    print(f"Validation runs directory {val_runs_dir} does not exist.")
```
